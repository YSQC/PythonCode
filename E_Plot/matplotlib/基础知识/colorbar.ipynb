```python
import  numpy as np
import  matplotlib.pyplot as plt

fig = plt.figure(figsize=(10, 7))
ax1 = plt.subplot(111)
x = np.random.randn(1000)
y = np.random.randn(1000)
color_list = list()
s = ax1.scatter(x, y, marker='v', alpha=0.9,
            s=np.abs(np.random.randn(1000)*100),
            c=x, cmap='Set1') # use the default colors.Normalize.(A class which, when called, can normalize data into the [0.0, 1.0] interval.)
ax1.grid()
fig.colorbar(s) # Add a colorbar to a plot.
fig.colorbar(s, orientation='horizontal', shrink=0.5) 
'''
orientation	vertical or horizontal
shrink	        1.0; fraction by which to multiply the size of the colorbar
aspect	        20; ratio of long to short dimensions
'''
```
