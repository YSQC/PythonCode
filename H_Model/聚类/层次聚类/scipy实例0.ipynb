```python
%matplotlib inline
from scipy.cluster.hierarchy import dendrogram, linkage # 使用scipy进行层次聚类
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

iris = load_iris()
X = iris.data
```

```python
# method='single' :最小距离
# method='complete' :最大距离
# method='average' :平均距离
# method='ward' :ward minimizes the variance of the clusters being merged.
Z = linkage(X, method='ward', 
            metric='euclidean') # 欧式距离
                                # cosine 与余弦距离
                                # correlation 相关系数
                                # chebyshev 切比雪夫距离
                                # cityblock 曼哈顿距离
fig = plt.figure(figsize=(25, 10))
dn = dendrogram(Z) # 分层聚类绘制为树状图
```

```python
X1 = [[i] for i in [2, 8, 0, 4, 1, 9, 9, 0]] # 矩阵形式
Z1 = linkage(X1, 'single')
dn1 = dendrogram(Z1)
```
